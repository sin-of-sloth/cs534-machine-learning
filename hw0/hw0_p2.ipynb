You recently cashed out your stock options, and now you have some money to invest. You are thinking of investing in the stock market, and you want to do it wisely. Being a data scientist, you believe historic patterns of stock prices may have some signals for future prices. First, you decide to analyze Apple (NASDAQ: AAPL).

1. Download the data from [here](https://yubin-park.github.io/cs534/homework/AAPL.csv), and plot the daily closing prices over time using Python.

```python
import pandas as pd
```

```python
data = pd.read_csv('./AAPL.csv')
data.plot(x='Date', y='Close')
```

2. Daily return is calculated as the ratio of consecutive closing prices (i.e. today’s closing price divided by yesterday’s closing price). From the data, obtain the mean and variance of the Apple stock’s daily return.

```python
data['Daily_Return'] = data['Close'] / data['Close'].shift(1)
print(f"Mean of Daily Returns : {data.mean(numeric_only=True)['Daily_Return']}")
print(f"Variance of Daily Returns : {data.var(numeric_only=True)['Daily_Return']}")
```

3. You think the daily return is a moving target, and you want to see the daily returns calculated in each year. Using the data, calculate the daily returns for the Year 2014, 2015, 2016, 2017, 2018, and 2019.

```python
for year in range(2015, 2021):
    yearly_data = data.loc[
        pd.DatetimeIndex(data['Date']).year == year
    ]
    print(f"Daily Return for the year {year} : {yearly_data.mean(numeric_only=True)['Daily_Return']}")
```
