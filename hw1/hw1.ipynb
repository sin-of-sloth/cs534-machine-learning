### Imports

````python
from math import sqrt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

### Utility Functions

````python
def load_energy_data(data_file_path):
    df = pd.read_csv(data_file_path, header=0)
    return df.iloc[:, 2:], df.iloc[:, 1]

"""
function to provide rmse and r^2 values for a prediction
given a model, x data, y data, and the dataset name
"""
def predict(model, x, y, dataset_name):
    y_pred = model.predict(x)
    print(
        '%10s | RMSE: %.10f | R^2: %.10f' % 
        (
            dataset_name,
            sqrt(mean_squared_error(y, y_pred)),
            r2_score(y, y_pred)
        )
    )

"""
function to do predictions given a model and a
dict of datasets
"""
def predict_all(model, datasets):
    for dataset_name, dataset in datasets.items():
        predict(model, dataset['x'], dataset['y'], dataset_name)
````

### Constants

````python
DATA_FOLDER = '.'
TRAINING_DATA = f'{DATA_FOLDER}/train.csv'
TEST_DATA = f'{DATA_FOLDER}/test.csv'
VALIDATION_DATA = f'{DATA_FOLDER}/validation.csv'

train_x, train_y = load_energy_data(TRAINING_DATA)
test_x, test_y = load_energy_data(TEST_DATA)
validation_x, validation_y = load_energy_data(VALIDATION_DATA)
datasets = {
    'train': {'x': train_x, 'y': train_y},
    'test': {'x': test_x, 'y': test_y},
    'validation': {'x': validation_x, 'y': validation_y}
}
````

````python
def FakeRidgeRegression():
    k = len(train_x.columns)
    lmbd = 99
    root_lmbd_i = [
        [
            sqrt(lmbd)
            if i == j
            else 0
            for j in list(range(k))
        ]
        for i in list(range(k))
    ]
    augmented_x_data = pd.concat([train_x, pd.DataFrame(root_lmbd_i, columns=train_x.columns)])
    augmented_y_data = pd.concat([train_y, pd.Series([0] * k)])

    linear_regr = linear_model.LinearRegression()
    linear_regr.fit(augmented_x_data, augmented_y_data)

    ridge_regr = linear_model.Ridge(alpha=lmbd)
    ridge_regr.fit(train_x, train_y)

    print('Linear Regr Coeffs\tRidge Regr Coeffs')
    for l, r in zip(linear_regr.coef_, ridge_regr.coef_):
        print('%f\t\t%f' % (l, r))

FakeRidgeRegression()


#############################################################################
#                                                                           #
#                                   P2                                      #
#                                                                           #
#############################################################################
#                           lambda = 99                                     #
#       =============================================================       #
#	        Linear Regr Coeffs	    Ridge Regr Coeffs						#
#           on augmented data                                               #
#	        	2.501970		    2.501965								#
#	        	-11.826618		    -11.826358								#
#	        	11.020954		    11.021202								#
#	        	-10.014662		    -10.015107								#
#	        	-18.148922		    -18.149181								#
#	        	21.570223		    21.570427								#
#	        	15.094566		    15.094241								#
#	        	-1.654536		    -1.654467								#
#	        	-1.682844		    -1.682780								#
#	        	-0.763359		    -0.763843								#
#	        	-0.229039		    -0.229014								#
#	        	5.129069		    5.128966								#
#	        	-0.045029		    -0.045076								#
#	        	0.475860		    0.475257								#
#	        	0.425812		    0.425930								#
#	        	2.256755		    2.257187								#
#	        	-4.419706		    -4.419757								#
#	        	-0.127176		    -0.128259								#
#	        	-5.391736		    -5.391586								#
#	        	-11.525292		    -11.526702								#
#	        	0.453376		    0.453388								#
#	        	-1.248509		    -1.248787								#
#	        	0.224808		    0.224639								#
#	        	0.231229		    0.231221								#
#	        	10.488749		    10.490268								#
#############################################################################
````

# P3: Modified Loss with Domain Knowledge

````python
def sgd(
    gradient, x, y, lmbd, n_vars=None, learn_rate=0.1, n_iter=10
):
    rng = np.random.default_rng(seed=None)
    coeff = np.ones(n_vars)
    coeff_prior = rng.normal(size=int(n_vars))
    for i in range(n_iter):
        for j in range(n_vars):
            weight_x = np.dot(x.iloc[j:j + 1, :], coeff)
            coeff_temp = (lmbd / n_vars) * np.dot(np.transpose(x.iloc[j:j + 1, :]), weight_x - y.iloc[j: j + 1])
            reg = lmbd * learn_rate * (coeff - coeff_prior) / n_vars
            coeff = coeff - coeff_temp - reg
    
    return coeff

def LinearRegressionWithKnowledge():
    def loss_func(X, y, coeff, coeff_prior, lmbd):
        res = (
            np.sum((np.dot(X, coeff) - y) * 2) + lmbd * (np.sum(coeff - coeff_prior) * 2)
        ) / 2 * len(X)
        return res

    alpha = 0.0000005
    custom_pen_coeffs = sgd(loss_func, train_x, train_y, alpha, n_vars=len(train_x.columns), learn_rate=0.000001)

    ridge_regr = linear_model.Ridge(alpha=alpha)
    ridge_regr.fit(train_x, train_y)

    print('Custom Penalty SGD Coeffs\tRidge Regr Coeffs')
    for l, r in zip(custom_pen_coeffs, ridge_regr.coef_):
        print('%f\t\t%f' % (l, r))
    

LinearRegressionWithKnowledge()

#############################################################################
#                                                                           #
#                                   P3                                      #
#                                                                           #
#############################################################################
#       Custom Penalty SGD Coeffs	Ridge Regr Coeffs                       #
#       0.919254		            2.491421                                #
#       0.953135		            -12.629751                              #
#       0.888752		            11.109823                               #
#       0.954653		            -10.533852                              #
#       0.895767		            -18.636192                              #
#       0.953947		            22.446800                               #
#       0.893739		            15.370178                               #
#       0.956152		            -1.431658                               #
#       0.891226		            -1.560615                               #
#       0.960149		            -0.641212                               #
#       0.860384		            -0.236222                               #
#       0.986033		            5.489064                                #
#       0.800825		            -0.072219                               #
#       0.959754		            1.079450                                #
#       0.901737		            0.398710                                #
#       0.957739		            2.468871                                #
#       0.886509		            -4.422324                               #
#       0.960884		            -1.033335                               #
#       0.895502		            -5.399325                               #
#       0.985972		            -18.335470                              #
#       -0.694622		            0.451701                                #
#       0.791822		            -2.421767                               #
#       0.986384		            0.129080                                #
#       0.898179		            0.234363                                #
#       0.989515		            17.185514                               #
#############################################################################
````
